```python
def hitung_total(harga, jumlah, diskon=0):
    total = harga * jumlah
    total -= total * (diskon / 100)
    return total

# Get user input (using predefined values for demonstration)
harga = 50000  # Example value
jumlah = 3     # Example value
diskon = 10    # Example value

# Calculate total using the function
total_belanja = hitung_total(harga, jumlah, diskon)

# Print the result matching the sample output format
print(f"Total belanja = {int(total_belanja)}")
```

```python
# Sample Output: (This was moved here as a comment for reference)
# Masukkan suhu Celcius: 30
# Suhu dalam Fahrenheit = 86.0

# File 1: konversi.py
def c_to_f(c):
    return (c * 9/5) + 32

def f_to_c(f):
    return (f - 32) * 5/9

# File 2: main.py
# In a real file system, you would import konversi from a separate file.
# For a Colab notebook, we define functions first and then use them.

# Using c_to_f function
c = float(input("Masukkan suhu Celcius: "))
print("Suhu dalam Fahrenheit =", c_to_f(c))

# Using f_to_c function
f = float(input("Masukkan suhu Fahrenheit: "))
print("Suhu dalam Celcius =", f_to_c(f))
```
